```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display all columns
pd.set_option("display.max_columns", None)

# Display all rows if needed
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully!")
```

```python
df1 = pd.read_csv("../data/raw/datafile.csv")
df2 = pd.read_csv("../data/raw/datafile (1).csv")
df3 = pd.read_csv("../data/raw/datafile (2).csv")
df4 = pd.read_csv("../data/raw/datafile (3).csv")
df5 = pd.read_csv("../data/raw/produce.csv")

print("All datasets loaded successfully!")
```

```python
# Check the dimensions of all datasets

print("Dataset 1:", df1.shape)
print("Dataset 2:", df2.shape)
print("Dataset 3:", df3.shape)
print("Dataset 4:", df4.shape)
print("Dataset 5:", df5.shape)
```

```python
# Display column names of each dataset

datasets = {
    "Dataset 1": df1,
    "Dataset 2": df2,
    "Dataset 3": df3,
    "Dataset 4": df4,
    "Dataset 5": df5
}

for name, df in datasets.items():
    print(f"\n{name}")
    print("-" * 50)
    print(df.columns.tolist())
```

```python
# Display the first 5 rows of each dataset

for name, df in datasets.items():
    print(f"\n{name}")
    display(df.head())
```

```python
# Display basic information about each dataset

for name, df in datasets.items():
    print(f"\n{'='*60}")
    print(name)
    print('='*60)
    df.info()
```

```python
# Check missing values

for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

```python
# Check duplicate rows

for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()} duplicate rows")
```

```python
# Statistical summary

for name, df in datasets.items():
    print(f"\n{name}")
    display(df.describe())
```
